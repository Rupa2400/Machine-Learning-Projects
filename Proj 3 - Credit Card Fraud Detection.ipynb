```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
creditcard_data = pd.read_csv('../ML PROJ/creditcard.csv')
```

```python
# Count for Class column
# 1--> Fraud Transactions 
# 0--> Legit Transactions

creditcard_data['Class'].value_counts()
```

```python
# Highly Unbalaced Dataset 

# Seperating all the data for analysis 

legit = creditcard_data[creditcard_data.Class == 0]
fraud = creditcard_data[creditcard_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical Measures of the data

# This is to understand the amount in both legit and fraud transactions 

print(legit.Amount.describe())

print(fraud.Amount.describe())
```

```python
# Comparing the values for both transactions

creditcard_data.groupby('Class').mean()
```

```python
# Under-Sampling

# Build a sample dataset containing similar distribution of Legit transactions and Fraudulent Transactions

# Number of Fraudulent Transactions --> 492 and Number of Legit Transactions --> 284315 

# So from Legit Transactions we'll take 492 values exactly the count which matches fraudulent transactions 
# create a new dataset merging these 2 values

# This will make it easier to predict the fraud pattern transactions
```

```python
legit_sample = legit.sample(n= 492)

legit_sample
```

```python
# Concatenating two datasets 

new_data = pd.concat([legit_sample, fraud], axis = 0)

# We mention axis = 0 while concatenation so all 492 values of fraud will be added below the legit_sample
# If we give axis = 1 then values will be added column wise which is not what we want
```

```python
new_data
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
X = new_data.drop(columns = 'Class', axis = 1)
Y = new_data['Class']
```

```python
X
```

```python
Y
```

```python
# Model Training

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
# Model

model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# Accuracy Score

X_train_prediction = model.predict(X_train)

training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print('Accuracy on training data : ', training_data_accuracy)
```

```python
# Accuracy Score

X_test_prediction = model.predict(X_test)

test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print('Accuracy on testing data : ', test_data_accuracy)
```
